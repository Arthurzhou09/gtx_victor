```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import os
import sys

project_root = '/home/victorh/projects/gtx'
os.chdir(project_root)
sys.path.insert(0, project_root)
```

```python
def plot(path, tf_or_py='tf'):

    log_df = pd.read_csv(path)

    if tf_or_py == 'tf':
        
        # a large figure with subplots
        fig, axs = plt.subplots(2, 3, figsize=(18, 8), sharex=True)

        # Row 1
        axs[0, 0].plot(log_df['epoch'], log_df['loss'], color='blue', label='Total Loss')
        axs[0, 0].set_title('Training Loss')
        axs[0, 0].legend()

        axs[0, 1].plot(log_df['epoch'], log_df['outDF_loss'], color='orange', label='outDF Loss')
        axs[0, 1].set_title('Output DF Loss')
        axs[0, 1].legend()

        axs[0, 2].plot(log_df['epoch'], log_df['outQF_loss'], color='green', label='outQF Loss')
        axs[0, 2].set_title('Output QF Loss')
        axs[0, 2].legend()

        # Row 2
        axs[1, 0].plot(log_df['epoch'], log_df['val_loss'], color='red', label='Validation Loss')
        axs[1, 0].set_title('Validation Total Loss')
        axs[1, 0].legend()

        axs[1, 1].plot(log_df['epoch'], log_df['val_outDF_loss'], color='purple', label='Validation DF Loss')
        axs[1, 1].set_title('Validation DF Loss')
        axs[1, 1].legend()

        axs[1, 2].plot(log_df['epoch'], log_df['val_outQF_loss'], color='brown', label='Validation QF Loss')
        axs[1, 2].set_title('Validation QF Loss')
        axs[1, 2].legend()

        for ax in axs[1, :]:
            ax.set_xlabel('Epoch')

        plt.tight_layout()
        plt.show()
        

plot('code_tf/ckpt/loss_20250516_100927.log', 'tf')
```
